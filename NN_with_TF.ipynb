```python
import numpy as np
from zipfile import ZipFile 
import pandas as pd
  
# loading the temp.zip and creating a zip object 
with ZipFile('/Users/paarthdudani/Downloads/connectionist+bench+sonar+mines+vs+rocks.zip', 'r') as zObject: 
    zObject.extractall() 

df = pd.read_csv('/Users/paarthdudani/Desktop/All folders/Python files/Miscellaneous_NN/sonar.all-data', header=None)

df[60].value_counts()
```

### Binary classifier

```python
# Extracting and processing the data 

dataset = df.to_numpy()

X = dataset[:,0:np.shape(dataset)[1]-1]
Y = dataset[:,-1]

# Binarizing the output.

for i in range(len(Y)):
    if  Y[i] == 'R':
        Y[i] = 1
    elif Y[i] == "M": 
        Y[i] = 0
```

### Building the NN

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25, random_state = 1)

X_train = tf.convert_to_tensor(X_train, dtype=tf.float32)
X_test = tf.convert_to_tensor(X_test, dtype=tf.float32)
Y_train = tf.convert_to_tensor(Y_train, dtype=tf.float32)
Y_test = tf.convert_to_tensor(Y_test, dtype=tf.float32)

n = 20
error_array = np.zeros((n),)
for i in range(n):
    model = keras.Sequential(
        [keras.layers.Dense(60, activation = 'relu'),
        keras.layers.Dense(20, activation = 'relu'),
        keras.layers.Dense(10, activation = 'relu'),
        keras.layers.Dense(1, activation = 'sigmoid')]
    )

    model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

    model.fit(X_train, Y_train, epochs = 20, batch_size = 8)

    model.evaluate(X_test,Y_test)

    Y_pred = model.predict(X_test).reshape(-1)
    Y_pred = np.round(Y_pred)

    def accuracy_calculator(y_pred, y_test):
        counts = np.zeros((len(y_pred)),)
        for i in range(len(y_pred)):
            if y_pred[i] != Y_test[i]:
                counts[i] = 1
            else:
                continue
        return 1- sum(counts)/len(counts)

    from sklearn.metrics import confusion_matrix, classification_report
    print(classification_report(Y_test, Y_pred))

    print()
    print(accuracy_calculator(Y_pred,Y_test))

    error_array[i] = accuracy_calculator(Y_pred,Y_test)


error_array
```
